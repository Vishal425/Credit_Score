```python
import cx_Oracle
import config
import pandas as pd
connection = None
try:
    connection = cx_Oracle.connect(
        config.username,
        config.password,
        config.dsn,
        encoding=config.encoding)

    # show the version of the Oracle Database
    print(connection.version)
    cur = connection.cursor()
    # query = 'select * from py_customer_all ca \
    #     inner join py_cust_general g on ca.py_customer_id=g.py_customer_id \
    #     inner join py_customer_mast m on ca.py_customer_id=m.py_customer_id\
    #     where ca.py_customer_id in (select py_customer_id from py_customer_mast where rownum < 1001)'
    query='select * from py_customer_all_loans al \
        inner join py_cust_general gl on al.py_customer_id=gl.py_customer_id \
        inner join py_customer_mast mt on gl.py_customer_id=mt.py_customer_id  where rownum < 54978'
    df = pd.read_sql_query(query, connection)
    print('data feach')
    df.to_csv('combine_1000_cust.csv',index=False)
    print('database save into csv') 
except cx_Oracle.Error as error:
    print('This is  the error:',error)
finally:
    # release the connection
    if connection:
        connection.close()
```
